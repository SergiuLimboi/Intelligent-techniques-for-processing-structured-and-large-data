# Assignment 2: Data Visualization for Large Structured Data

## Deadline: week 3.  Each delay is penalized with 3 points. Latest deadline: week 5.

## The goal of this assignment is to practice responsible data visualization for large datasets.

Each student must choose one dataset from the list below. Alternatively, they may select a dataset of their own, provided that it is structured, available in Parquet format, and contains at least 3 million records.

All datasets:
* are large
* are structured
* come from real-world sources
* are suitable for aggregation and visualization

### Chicago Taxi Trips
Source: City of Chicago--- https://data.cityofchicago.org/Transportation/Taxi-Trips/wrvz-psew

### U.S. Airline On-Time Performance
Source: U.S. Department of Transportation--- https://www.transtats.bts.gov/ONTIME/


### UK Road Traffic Accidents
Source: UK Department for Transport--- https://www.kaggle.com/datasets/tsiaras/uk-road-safety-accidents-and-vehicles

### NYC Weather Observations (NOAA via AWS)
Parquet format: "s3://noaa-ghcn-pds/parquet/by_year/2023.parquet"





## Rules:
* You must not visualize raw millions of rows directly.
* Sampling is allowed only if justified.
* Aggregation is mandatory for all plots.
* CSV or Parquet formats are both accepted (Parquet preferred).

## Task 1. Data Loading (1 point)

Load the dataset using Polars
* Prefer lazy loading (scan_csv, scan_parquet)
* Show dataset size (number of rows & columns)

## Task 2. Exploratory Data Analysis (EDA) (2 points)
Show:
* Missing values per column
* Basic statistics for key numeric variables
* Identification of  outliers

## Task 3. Data Vizualization (7 points)

You must include at least 7 visualizations, covering:

* Distribution plot (e.g. histogram, boxplot)
* Categorical aggregation (e.g. average / count per category)
* Time-based visualization (e.g. per day, per hour, per month)
* Relationship visualization (e.g. hexbin, aggregated scatter)
